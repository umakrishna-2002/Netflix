```python
import pandas as pd

# Step 1: Clean and prepare
df_clean = df.dropna(subset=['Title', 'Country', 'Category']).copy()
df_clean['Country'] = df_clean['Country'].str.split(',')
df_clean['Country'] = df_clean['Country'].apply(lambda x: [c.strip() for c in x])
df_exploded = df_clean.explode('Country')

# Step 2: Group by Title and Category for country count
title_country_counts = (
    df_exploded.groupby(['Title', 'Category'])['Country']
    .nunique()
    .reset_index(name='Country_Count')
)

# Step 3: Find the top Movie and TV Show
top_movie_title = title_country_counts[title_country_counts['Category'] == 'Movie'].sort_values(by='Country_Count', ascending=False).iloc[0]['Title']
top_tv_title = title_country_counts[title_country_counts['Category'] == 'TV Show'].sort_values(by='Country_Count', ascending=False).iloc[0]['Title']

# Step 4: Get country lists
movie_countries = df_exploded[df_exploded['Title'] == top_movie_title]['Country'].unique().tolist()
tv_countries = df_exploded[df_exploded['Title'] == top_tv_title]['Country'].unique().tolist()

# Step 5: Build final table
final_table = pd.DataFrame({
    'Title': [top_movie_title, top_tv_title],
    'Category': ['Movie', 'TV Show'],
    'Country_Count': [len(movie_countries), len(tv_countries)],
    'Countries': [', '.join(sorted(movie_countries)), ', '.join(sorted(tv_countries))]
})

# Display the table
print("Most Widely Released Movie and TV Show based on it released in most countires:")
display(final_table)
```

```python
import pandas as pd

# Drop missing values in required columns
df_cleaned = df.dropna(subset=['Country', 'Rating', 'Category']).copy()

# Explode comma-separated countries into separate rows
df_cleaned.loc[:,'Country'] = df_cleaned['Country'].str.split(',')
df_exploded = df_cleaned.explode('Country')
df_exploded['Country'] = df_exploded['Country'].str.strip()

# Group by Country, Category, and Rating
grouped = df_exploded.groupby(['Country', 'Category', 'Rating']).size().reset_index(name='No_of_Releases')

# Sort so highest release rating is on top
sorted_grouped = grouped.sort_values(['Country', 'Category', 'No_of_Releases'], ascending=[True, True, False])

# Get the most released rating per Country & Category
most_released = sorted_grouped.groupby(['Country', 'Category']).first().reset_index()

# Rename for clarity
most_released = most_released.rename(columns={'Rating': 'Most_Released_Rating'})

# Separate tables
movies_df = most_released[most_released['Category'] == 'Movie'].sort_values(by='No_of_Releases', ascending=False).reset_index(drop=True)
tvshows_df = most_released[most_released['Category'] == 'TV Show'].sort_values(by='No_of_Releases', ascending=False).reset_index(drop=True)

# Display the sorted tables
print(" Most Released Rated Movies by Country (Highest to Lowest):")
display(movies_df)

```

```python
import matplotlib.pyplot as plt

# Sort and take top 10 countries
top_movies_df = movies_df.sort_values(by='No_of_Releases', ascending=False).head(10)

# Horizontal Bar Chart
plt.figure(figsize=(10, 6))
bars = plt.barh(top_movies_df['Country'], top_movies_df['No_of_Releases'], color='skyblue')
plt.xlabel('Number of Releases')
plt.title('🎬 Top 10 Countries by Number of Released Rated Movies')

# Add text annotations for ratings
for bar, rating in zip(bars, top_movies_df['Most_Released_Rating']):
    plt.text(bar.get_width() + 1, bar.get_y() + bar.get_height()/2, rating, va='center')

plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Drop missing values
df_cleaned = df.dropna(subset=['Country', 'Category', 'Type']).copy()

# Explode countries into rows (in case of multiple countries)
df_cleaned.loc[:,'Country'] = df_cleaned['Country'].str.split(',')
df_exploded = df_cleaned.explode('Country')
df_exploded['Country'] = df_exploded['Country'].str.strip()

# Group by Country, Category (Movie/TV Show), and Type (Genre)
grouped = df_exploded.groupby(['Country', 'Category', 'Type']).size().reset_index(name='Count')

# Sort to get the most released/watched Type per country and category
sorted_grouped = grouped.sort_values(['Country', 'Category', 'Count'], ascending=[True, True, False])

# Keep only the top Type per Country and Category
top_type_by_country = sorted_grouped.groupby(['Country', 'Category']).first().reset_index()

# Rename for clarity
top_type_by_country = top_type_by_country.rename(columns={'Type': 'Most_Popular_Type'})

# Show top countries for Movies and TV Shows separately
top_movies_type = top_type_by_country[top_type_by_country['Category'] == 'Movie'].sort_values(by='Count', ascending=False).reset_index(drop=True)
top_shows_type = top_type_by_country[top_type_by_country['Category'] == 'TV Show'].sort_values(by='Count', ascending=False).reset_index(drop=True)

# Display the results
print(" Most Popular Movie Types by Country:")
display(top_movies_type)


```

```python
from textblob import TextBlob
# Drop NaN descriptions
df_clean = df.dropna(subset=['Description'])

# Apply sentiment analysis
df_clean['Polarity'] = df_clean['Description'].apply(lambda x: TextBlob(x).sentiment.polarity)
df_clean['Sentiment'] = df_clean['Polarity'].apply(
    lambda x: 'Positive' if x > 0 else ('Negative' if x < 0 else 'Neutral')
)
```

```python
print(df_clean['Sentiment'].value_counts())
```

```python
import matplotlib.pyplot as plt

df_clean['Sentiment'].value_counts().plot.pie(autopct='%1.1f%%', startangle=140)
plt.title('Sentiment Distribution of Netflix Descriptions')
plt.ylabel('')
plt.show()
```

```python
# Top 5 Positive
print(df_clean[df_clean['Sentiment'] == 'Positive'][['Title', 'Description']].head())

# Top 5 Negative
print(df_clean[df_clean['Sentiment'] == 'Negative'][['Title', 'Description']].head())
```

```python
# Filter only Comedy and Drama
comedy_drama_df = df_clean[df_clean['Type'].isin(['Comedies', 'Dramas'])]

# Group by Type and calculate average polarity
avg_sentiment = comedy_drama_df.groupby('Type')['Polarity'].mean().reset_index().sort_values(by='Polarity', ascending=False)

import matplotlib.pyplot as plt

# Plot
plt.figure(figsize=(6,4))
plt.barh(avg_sentiment['Type'], avg_sentiment['Polarity'])
plt.xlabel("Average Sentiment Polarity")
plt.title("Are Comedies More Positive Than Dramas?")
plt.grid(True)
plt.show()

avg_sentiment
```
